```python
from langchain_community.document_loaders import PDFPlumberLoader
from langchain_experimental.text_splitter import SemanticChunker
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain
from langchain.chains.combine_documents.stuff import StuffDocumentsChain
from langchain.chains import RetrievalQA
```

```python
loader=PDFPlumberLoader("/content/neuralink-technology-the-future-of-neural-engineering.pdf")
docs=loader.load()
```

```python
docs
```

```python
text_splitter=SemanticChunker(HuggingFaceEmbeddings())
documents=text_splitter.split_documents(docs)
```

```python
embedder=HuggingFaceEmbeddings()
```

```python
vector=FAISS.from_documents(documents,embedder)
retriever=vector.as_retriever(search_type="similarity",search_kwargs={"k":3})
```

```python
#define prompt
prompt = """
1. Use the following piece of context to answer the question at the end.
2. If you don't know the answer, just say that "I don't know" but do not make up an answer on your own .\n
3. Keep the answer crisp and limited to 3–4 sentences.

Context: {context}
Question: {question}
Helpful Answer:
"""

QA_CHAIN_PROMPT = PromptTemplate.from_template(prompt)
```

```python
llm_chain=LLMChain(
    llm=llm,
    prompt=QA_CHAIN_PROMPT,
    callbacks=None,
    verbose=True
)
```

```python
document_prompt=PromptTemplate(
    input_variables=["Page_content","source"],
    template="context: \ncontent:{page_content}\nsource;{source}"
)
```

```python
combine_documents_chain=StuffDocumentsChain(
    llm_chain=llm_chain,
    document_variable_name="context",
    document_prompt=document_prompt,
    callbacks=None

)
```

```python
qa=RetrievalQA(combine_documents_chain=combine_documents_chain,
               verbose=True,
               retriever=retriever,
               return_source_documents=True)
```

```python
 user_input="what is literature review?"
 response=qa(user_input)["result"]
```
